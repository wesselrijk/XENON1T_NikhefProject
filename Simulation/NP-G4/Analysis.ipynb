Import the necessary packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import the data

```python
_data = np.genfromtxt(fname = ".//edep.csv", delimiter = ",")
```

Make the projections

```python
_LxyE = np.zeros((20,20))
_LyzE = np.zeros((20,20))
for s in _data:
    _LxyE[int(s[0]),int(s[1])] += s[3]
    _LyzE[int(s[1]),int(s[2])] += s[3]
```

Plot the projections

```python
#print(_Lx)
_xy = sns.heatmap(_LxyE)
plt.show()
plt.close()
_yz = sns.heatmap(_LyzE)
plt.show()
```
